```python
dataset = pd.read_csv("verim.csv",encoding = 'ISO-8859-1')
```

```python
dataset.head
```

```python
one_hot = OneHotEncoder(handle_unknown='ignore')
one_hot_cinsiyet = one_hot.fit_transform(dataset['Cinsiyet'].values.reshape(-1,1)).toarray()
one_hot_cinsiyet
```

```python
one_hot.categories_
```

```python
one_hot_df = pd.DataFrame(one_hot_cinsiyet, columns = one_hot.categories_)
one_hot_df.head()
```

```python
one_hot_dfjoin = dataset.join(one_hot_df)
one_hot_dfjoin.head()
```

```python
one_hot_dfjoin.drop("Cinsiyet", axis=1, inplace= True)
one_hot_dfjoin.head()
```

```python
X = one_hot_dfjoin.drop("AlimDurumu",axis=1)
y=  one_hot_dfjoin["AlimDurumu"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size =0.3,random_state= 1)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
xx= MinMaxScaler()
x_train = xx.fit_transform(x_train)
x_test = xx.transform(x_test)
```

```python
x_train
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_true=y_test, y_pred =y_pred)
```

```python
print(classification_report (y_true=y_test, y_pred = y_pred))
```

```python
print ("dogruluk  değeri:", accuracy_score (y_true=y_test, y_pred = y_pred))
```
